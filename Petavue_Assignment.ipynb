# Dealing with flight data

```python
import os

base_folder_path = '/content/drive/MyDrive/Petavue/flightdata2016'

# List of airport codes
airport_codes = ['ATL', 'CLT', 'DEN', 'DFW', 'EWR', 'IAH', 'JFK', 'LAS', 'LAX', 'MCO', 'MIA', 'ORD', 'PHX', 'SEA', 'SFO']

# recommended flight columns
recommended_columns = ['FlightDate', 'Quarter', 'Year', 'Month', 'DayofMonth', 'DepTime', 'DepDel15', 'CRSDepTime',
                        'DepDelayMinutes', 'Origin', 'Dest', 'ArrTime', 'CRSArrTime', 'ArrDel15', 'ArrDelayMinutes']


combined_df = pd.DataFrame()

for month in range(1, 13):

    file_path = os.path.join(base_folder_path, f'On_Time_On_Time_Performance_2016_{month}', f'On_Time_On_Time_Performance_2016_{month}.csv')

    df = pd.read_csv(file_path)

    # filter data based on airport codes
    filtered_df = df[df['Origin'].isin(airport_codes) | df['Dest'].isin(airport_codes)]

    # select recommended flight columns
    final_df = filtered_df[recommended_columns]

    # concatenate the current DataFrame with the combined DataFrame
    combined_df = pd.concat([combined_df, final_df], ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('filtered_2016_data.csv', index=False)
```

# Weather dataset preprocessing


```python
main_directory = "/content/drive/MyDrive/Petavue/weather"
```

```python
import json
```

```python
def parse_json_to_dataframe(json_data):
    hourly_data = []

    for day in json_data['data']['weather']:
        for hour in day['hourly']:
            hourly_data.append({
                'time': int(hour['time']),
                'tempF': float(hour['tempF']),
                'humidity': int(hour['humidity']),
                'windSpeedKmph': int(hour['windspeedKmph']),
                'cloudcover': int(hour['cloudcover']),
                'precipMM': float(hour['precipMM']),
                'visibility': int(hour['visibility']),
                'weatherCode': int(hour['weatherCode']),
                'airport': airline_folder,
                'WindGustKmph' : int(hour['WindGustKmph']),
                'WindChillF' : int(hour['WindChillF']),
                'Pressure' : int(hour['pressure']),
                'DewPointF' : int(hour['DewPointF']),
                'WindDirDegree' : int(hour['winddirDegree']),
                'date' : str(day['date']),
            })

    data = pd.DataFrame(hourly_data)
    return data
```

```python
main_directory = "/content/drive/MyDrive/Petavue/weather"
dataframes = []  # to store DataFrames for each month

for airline_folder in os.listdir(main_directory):
    airline_path = os.path.join(main_directory, airline_folder)

    for month_file in os.listdir(airline_path):
        month_path = os.path.join(airline_path, month_file)
        year = int(month_file.split('-')[0])

        if year in [2016, 2017]:
            with open(month_path, 'r') as file:
                month_data = json.load(file)

            df = parse_json_to_dataframe(month_data)
            dataframes.append(df)
```

```python
final_df = pd.concat(dataframes, ignore_index=True)
```
